```python
import numpy as np
```

```python
y_value_array = [-0.009784886185544616,  -0.010898639041610531, -0.011184225824071402]
```

```python
# 1 -> Finest Mesh
# 3 -> Coarsest Mesh
 
f_3 = y_value_array[0]
f_2 = y_value_array[1]
f_1 = y_value_array[2]
```

```python
(np.log((f_3 - f_2)/(f_2 - f_1)))/np.log(2)
```

```python
ans = -0.01145667
# ans = -0.0112

mesh_spacing_array = [2/25, 2/50, 2/100]
deflection_error = [(ans - y_value_array[0])/ans, (ans - y_value_array[1])/ans, (ans - y_value_array[2])/ans] 
```

```python
from matplotlib import pyplot as plt

plt.scatter(mesh_spacing_array, deflection_error)
plt.plot(mesh_spacing_array, deflection_error)
plt.xscale("log")
plt.yscale("log")
```
