```python
ans = - 0.01456
# ans = -0.0112

mesh_spacing_array = [2/25, 2/50, 2/100, 2/200]
deflection_error_python = [(ans - y_value_array_python[0])/ans, (ans - y_value_array_python[1])/ans, (ans - y_value_array_python[2])/ans, (ans - y_value_array_python[3])/ans] 

deflection_error_s4f = [(ans - y_value_array_s4f[0])/ans, (ans - y_value_array_s4f[1])/ans, (ans - y_value_array_s4f[2])/ans, (ans - y_value_array_s4f[3])/ans] 
```

```python
from matplotlib import pyplot as plt

plt.scatter(mesh_spacing_array, deflection_error_python)
plt.plot(mesh_spacing_array, deflection_error_python, label = "python")
plt.scatter(mesh_spacing_array, deflection_error_s4f)
plt.plot(mesh_spacing_array, deflection_error_s4f, label = "s4f")
plt.legend()
plt.xscale("log")
plt.yscale("log")
```
