```python
y_value_array_python = [-1.621881575912283e-05, -1.7790274630665906e-05, -1.8261187112413966e-05]
y_value_array_s4f = [-1.68122e-05, -1.80125e-05, -1.83312e-05]
```

```python
ans = -1.89e-05

mesh_spacing_array = [2/50, 2/100, 2/200]


deflection_error_python = [(ans - y_value_array_python[0])/ans, (ans - y_value_array_python[1])/ans, (ans - y_value_array_python[2])/ans] #, (ans - y_value_array_python[3])/ans , (ans - y_value_array_python[4])/ans] 

deflection_error_s4f = [(ans - y_value_array_s4f[0])/ans, (ans - y_value_array_s4f[1])/ans, (ans - y_value_array_s4f[2])/ans] #, (ans - y_value_array_s4f[3])/ans] 
```

```python
from matplotlib import pyplot as plt

plt.scatter(mesh_spacing_array, deflection_error_python)
plt.plot(mesh_spacing_array, deflection_error_python, label = "python")
plt.scatter(mesh_spacing_array, deflection_error_s4f)
plt.plot(mesh_spacing_array, deflection_error_s4f, label = "s4f")
plt.legend()
plt.xscale("log")
plt.yscale("log")
```
